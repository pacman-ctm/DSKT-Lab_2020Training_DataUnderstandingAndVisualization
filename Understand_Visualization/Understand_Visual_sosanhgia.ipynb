```python
import pandas as pd
import matplotlib.pyplot as plt   
import matplotlib.ticker as ticker
import numpy as np
import seaborn as sns
```

```python
def changeTextToInt(text):
    # Convert text "2.300.000" to int 2300000
    if (type(text) == str):
        return int(float(text.replace(".","")))
    return np.nan

    
```

```python
df = pd.read_csv("sosanhgia.csv")
df
```

```python
pd.DataFrame(df, columns = ['category__001', 'name', 'brand', 'priority_price', 'priority_store', 'num_of_product', 'num_of_store', 'lowest_price', 'highest_price'] )
```

```python
#df['priority_price'] = pd.to_numeric(df['priority_price'],errors='coerce')
df
```

```python
df.dtypes
```

```python
# Change the object -> float64
df['priority_price'] = df['priority_price'].apply(changeTextToInt)
df['lowest_price'] = df['lowest_price'].apply(changeTextToInt)
df['highest_price'] = df['highest_price'].apply(changeTextToInt)
```
